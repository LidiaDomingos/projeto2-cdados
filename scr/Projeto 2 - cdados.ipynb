#### O que é câncer de mama?

>>"*Câncer de mama é o tipo de câncer mais comum entre as mulheres no mundo e no Brasil, depois do câncer de pele não melanoma. O câncer de mama responde, atualmente, por cerca de 28% dos casos novos de câncer em mulheres. O câncer de mama também acomete homens, porém é raro, representando menos de 1% do total de casos da doença. Relativamente raro antes dos 35 anos, acima desta idade sua incidência cresce progressivamente, especialmente após os 50 anos. Estatísticas indicam aumento da sua incidência tanto nos países desenvolvidos quanto nos em desenvolvimento. Existem vários tipos de câncer de mama. Alguns evoluem de forma rápida, outros, não. A maioria dos casos tem bom prognóstico. Para o ano de 2018, foram estimados 59.700 novos casos de câncer de mama no Brasil.* 


#### Quais sintomas do câncer de mama?

>>*O sintoma mais comum de câncer de mama é o aparecimento de nódulo, geralmente indolor, duro e irregular, mas há tumores que são de consistência branda, globosos e bem definidos. Outros sinais de câncer de mama são:*
>>* edema cutâneo (na pele), semelhante à casca de laranja;
* retração cutânea;
* dor;
* inversão do mamilo;
* hiperemia;
* descamação ou ulceração do mamilo;
* secreção papilar, especialmente quando é unilateral e espontânea.

>>*A secreção associada ao câncer geralmente é transparente, podendo ser rosada ou avermelhada devido à presença de glóbulos vermelhos. Podem também surgir linfonodos palpáveis na axila. A postura atenta das mulheres em relação à saúde das mamas, que significa conhecer o que é normal em seu corpo e quais as alterações consideradas suspeitas de câncer de mama, é fundamental para a detecção precoce dessa doença.*

>>*Sinais e sintomas devem sempre ser investigados, pois podem estar relacionados a doenças benignas ou malignas da mama.*
    
#### Tumor benigno
 
>>*Tem células que crescem lentamente e semelhante às do tecido normal. Na maioria dos casos pode ser totalmente removido (e o paciente curado) por meio de cirurgia.*

#### Tumor maligno

>>*As células multiplicam-se rapidamente e têm a capacidade de “invadir” estruturas próximas ao local de origem. A cura neste tipo de tumor depende do diagnóstico precoce e do tratamento adotado.
    A classificação entre tumores benignos e malignos é feita com base em três pontos principais, que são:*

#### Melhorando e separando os dados para serem usados nas Técnicas de Previsão
Dizemos que é uma classificação quando o nosso objetivo é classificar um caso dentre algumas opções. Nesse sentido, é preciso criar um dataframe apenas com as features, pois são os dados que vão ser observados para definir se é o tumor é benigno ou maligno. Para isso, é criado uma nova base de dados sem os diagnósticos, chamada dados explicativos.

```python
dados = pd.read_csv(filename) #colocado aqui apenas para não dar erro caso rodar essa célula de novo
del dados['diagnosis']
del dados['Unnamed: 32']
del dados['id']
dados_explicativos = dados
dados_explicativos
```

Ademais, para obter uma previsão mais precisa, é necessário que ocorra uma padronização dos dados. Grande parte dos estimadores de machine learning do scikit-learn podem ter uma aficácia menor quando não é utilizado dados que estejam parecidos com o formato de média zero e variância, pois os modelos scikit-learn já são produzidos levando em conta que será usado dados que estejam próximos do modelo de distribuição normal padronizada.

```python
dados_explicativos_ajustados = StandardScaler().fit_transform(dados_explicativos)

resultados_ajustados = LabelEncoder().fit_transform(dados_limpos.diagnosis)
```

Também é possível, por meio da tabela abaixo, confirmar, na padronização feita acima, o que é dado como Benigno se transforma no número 0 e o que é Maligno se torna 1.

```python
resultados_em_B_e_M = pd.Series(dados_limpos.diagnosis).array
df = pd.DataFrame()
df['resultado em B e M'] = resultados_em_B_e_M
df['resultado em 0 e 1'] = resultados_ajustados
df
```

Antes de começar com as técnicas, também é necessário separar a base de dados entre treinamento e teste, pois é necessário treinar o classificador primeiro e depois fazer o teste, e fazer a comparação dos valores previstos com os reais.

```python
X_train, X_test, y_train, y_test = train_test_split(dados_explicativos_ajustados, resultados_ajustados, test_size = 0.3, random_state=42)
f'Tamanho do dataframe de treinamento = {len(X_train)} e tamanho do dataframe de teste = {len(X_test)}'
```

## Primeira Técnica de Classificação - Método da Regressão Logística

> Utilizada em problemas que envolvam a classificação de certos dados, o método da Regressão Logística tem um objetivo 
análoga à regressão linear múltipla, porém seu resultado é dado de forma binária, ou seja, com duas opções opostas. Essa é uma das características da técnica, em que uma parte da separação dos dados serve para estimação do modelo, e outra parte para testar a eficiência da classificação. Além disso, utiliza-se de variáveis dependentes e tem por objetivo a determinação da probabilidade de ocorrência dos valores que se quer prever acerca de uma variável qualitativa.
> De maneira resumida, a regressão logística analisa dados distribuídos binomialmente e com ela se consiguirá prever a probabilidade do tumor ser benigno ou maligno com base nas várias features selecionadas.

Vantagens do Modelo Logístico segundo a [referência 15]:
- Facilidade para lidar com variáveis independentes categóricas.
- Fornece resultados em termos de probabilidade.
- Facilidade de classificação de indivíduos em categorias.
- Requer pequeno número de suposições.
- Alto grau de confiabilidade.


Inicialmente, é necessário chamar a função de Regressão Logística na biblioteca scikit-learn

```python
classificador_logistico = LogisticRegression()
```

Depois, é necessário treinar o modelo com a função .fit(), a qual se ajusta para poder dar a melhor previsão:

```python
classificador_logistico.fit(X_train,y_train)
```

Após o modelo ter sido treinado, apenas precisa fazer a previsão para a base de teste das features, com a função predict()

```python
ypred = classificador_logistico.predict(X_test)
ypred
```

É possível ver um relatório geral do modelo brevemente, sendo aprofundados mais a frente:

```python
#Relatório da  1° técnica
print(classification_report(y_test, ypred))
```

### Feature Importance
É possível observar também quais as variáveis que são mais importantes para o modelo de Regressão Logística, quanto maior o coeficiente, melhor para o modelo essa variável é. Ou seja: quanto maior o valor, mais importante a feature é para o modelo.

```python
coefs = np.abs(classificador_logistico.coef_[0])
indices = np.argsort(coefs)[::-1]
plt.figure()
plt.title("Importância de cada feature - Regressão Logística")
plt.bar(range(10), coefs[indices[:10]],
       color="violet", align="center")
plt.xticks(range(10), dados_explicativos.columns[indices[:10]], rotation=45, ha='right')
plt.subplots_adjust(bottom=0.3)
```

## Segunda Técnica de Classificação - Método da Árvore de Decisão

>O método da árvore de decisão é utilizado para responder uma pergunta que terá como opção de resposta um sim ou não, por exemplo. Analogamente com programação, é como se fosse um conjunto de declarações if-else, e que quando aplicados a uma amostra de dados, resulta em sua classificação. Nela, o nó raiz é o atributo que melhor divide o conjunto de dados. Nós de decisão são representados como condições lógicas que determinam o caminho dentro da árvore. Branch/Sub árvore, são árvores menores no meio do caminho que surgem a partir do nó de decisão e o nó terminal/folha estão na parte inferior da árvore e pode conter uma classe ou um valor contínuo. Como a árvore está sendo utilizada para classificação, os critérios de partição mais conhecidos são baseados na entropia e índice Gini, ao final vamos conseguir responder se o tumor será beligno ou maligno.



Exemplo de uma árvore abaixo: os nós são representados pelos quadros com as perguntas e as informações podem ser consideradas as perguntas e suas possíveis respostas.

### 1. Estatísticas da Regressão Logística

>Um dos primeiros termos que podem ser observados é a acurácia do modelo, ou seja, o quanto ele acerta quando recebe as features e os rótulos esperados. Esse termo pode variar de 0 até 1, sendo que quando mais próximo de 1, melhor é a eficácia do modelo. Além disso, é também importante observar se não ocorre o 'overfitting', que nada mais seria do o modelo funcionar muito bem para a base de dados de treinamento, porém ter baixo score para a base de teste. Isso indica quando o modelo não tem muita qualidade, quando recebe um conjunto de dados nunca visto.

```python
#Acurácia do modelo para a base de treinamento
score_train_reglog = classificador_logistico.score(X_train, y_train)
score_test_reglog = classificador_logistico.score(X_test, y_test)
f'A acurácia para a base de treinamento é {score_train_reglog:.4f} e a acurácia para a base de teste é {score_test_reglog:.4f}'
```

>Como a diferença entre os dois scores é bem pequena, então é comprovado que não ocorre o 'overfitting', ou seja: o modelo está funcionando bem para as duas bases! É também possível ver como o modelo está funcionando em questão de acertos e erros com uma matriz de confusão. Usando o projeto aqui descrito como exemplo, a tabela apresenta a quantidade de verdadeiros positivos (confirmando que o modelo prevê B e a resposta é de fato B (benigno)), falsos positivos (prevê M (maligno), mas a resposta é B (benigno)), verdadeiros negativos (prevê M (maligno) e a resposta é de fato M (maligno) e falsos negativos (prevê que seja B (benigno) e a resposta é M (maligno)).

> Para uma melhor compreensão dos resultados, é possível fazer a **Validação Cruzada**, pra ver como o modelo funciona de forma generalizada. Nesse sentido, um das formas de fazer essa validação é com o método de K-Fold, o qual recebe a base de dados do projeto, sendo dividida em K vezes, usando K-1 como treino e 1 como teste, ou seja, melhorando o treinamento do modelo, obtendo um score mais preciso, a média e o desvio-padrão, assim observando como o modelo de comporta de forma mais precisa. É possível entender melhor o método com a imagem abaixo:

<img src="img/kfold.png" width="500" width="50%" >
fonte: [Referencia 17]

R² é o percentual de variação na resposta que é prevista pelo modelo, quanto mais próximo de 1, melhor, pois significa que o modelo se adequa bem a diferentes conjuntos de dados. 

```python
x = dados_explicativos_ajustados
y = resultados_ajustados
kfold  = KFold(n_splits=10, shuffle=True) # shuffle=True, Shuffle (embaralhar) the data.
result = cross_val_score(classificador_logistico, x, y, cv = kfold)

for i in result:
    print("(R²) Score : {0}".format(i))
print("Média de R²: {0}".format(result.mean()))
print("Desvio padrão de R²: {0}".format(result.std()))
```

Percebe-se que o desvio padrão tem um valor considerável, em torno de 2% de variação, mostrando que o modelo tem uma certa mudança quando são diferentes conjuntos de dados, porém, ainda assim, o modelo apresenta uma **boa acurácia** pela média maior que 97%.
